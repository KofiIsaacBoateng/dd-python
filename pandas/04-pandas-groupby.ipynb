# Group By

The groupby method allows you to group rows of data together and call aggregate functions

```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['Finance', 'Marketing', 'Marketing', 'Sales', 'Finance', 'Finance', 'Sales', 'Finance'],
    'Employee': ['Sam', 'James', 'William', 'Jessica', 'Roland', 'Troy', 'Tracy', 'Ammy'],
    'Rank': [2, 3, 2, 1, 5, 3, 5, 4]
})
```

```python
df
```

**Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:**

```python
df.groupby("Department")
```

You can save this object as a variable and then call aggregate methods \
off the object

```python
grouped = df.groupby('Department')
grouped.mean("Rank")
```

```python
grouped.min()
```

```python
grouped.max()
```

```python
grouped.count()
```

```python
grouped.describe()
```

```python
grouped.describe().transpose()
```

```python
grouped.describe().transpose()[['Finance', 'Marketing']]
```

```python
grouped.describe().transpose().loc['Rank']
```
